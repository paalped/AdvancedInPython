If we modify our function numerFilter to make it more generic, we can reuse this code.

But first lets create a small decorator to help us, if you have seen my previous course an introduction to python decorators, this will make sence. 

```python
def filterDecorator(function_input, accept_this):
    kwargs = locals() # in this local scope only the above variable names will exist
    # since locals is a dict we can use it as our kwargs
    def wrapper(function):
        def _wrapper(*args, **kw):
            # If we had called kw for kwargs, we would have overwritten kwargs from outer scope
            # we need to extend kwargs with kw
            #kwargs = {**kwargs , **kw} # <-- fastest way to extend two dicts in Python 3
            # Doesn't work in this case :/
            for k, v in kw.items():
                kwargs[k] = v
            return function(*args, **kwargs)
        return _wrapper
    return wrapper
```

```python
@filterDecorator(function_input=0, accept_this=False)
def genericFilter(_tuple_items, **kwargs):
    keyword, function = _tuple_items
    function_input = kwargs['function_input']
    accept_this = kwargs['accept_this']
    if keyword in ('input', 'copyright', 'credits', 'help', 'print', 'display', 'license', 'dir'):
        # We don't wanna try to evaulate these, cause they cause anoying outputs
        return False
    try:
        function(function_input)
        return accept_this
    except:
        return not accept_this
```

So we created this decorator so it can store som values which we can use together with the filter function. <br/>
If we had not done this, our filter function would have to rely on some hardcoded parameters, but we don't want that.

While this works its not really pretty to recreate a whole new function, just to change a small little thing. Lets try and create an object to make it more clean and use a nicer syntax.

```python
class Filter(object):
    def __init__(self, finput, accept: bool):
        self.finput = finput
        self.accept = accept
    def __call__(self, arg):
        keyword, function = arg
        if keyword in ('input', 'copyright', 'credits', 'help', 'print', 'display', 'license', 'dir'):
            # We don't wanna try to evaulate these, cause they cause anoying outputs
            return False
        try:
            function(self.finput)
            return self.accept
        except:
            return not self.accept
```

```python
accepts_string_numbers = filter(Filter('123', True), fewer_builtins.items())
```

```python
for kw, func in accepts_string_numbers:
    print(f'{kw}: {kw}--> {func("123")}')
```

```python
class FilterCallableKeywords(object):
    def __init__(self, *args, accept=True):
        self.args = args
        self.accept = accept
    def __call__(self, keyword):
        if keyword in ('input', 'copyright', 'credits', 'help', 'print', 'display', 'license', 'dir'):
            # We don't wanna try to evaulate these, cause they cause anoying outputs
            return False
        try:
            function = eval(keyword)
            function(*self.args)
            return self.accept
        except:
            return not self.accept
```

```python
class CallableKeywords(object):
    def __init__(self, *args, accept=True):
        self.args = args
        self.accept = accept
        self.keyword = 'Not initialized'
        self.function = 'Not initialized'
        self.error = 'Not raised'
        self.result = None
        self.name = self.init_name()
    
    @classmethod
    def init_name(cls):
        __doc__ = 'A method to get the class name'
        return cls.__name__
    
    def __call__(self, keyword: str):
        __doc__ = '''Evaluates a keyword, into a function. stores True or False if its 
                    a valid function, based on the init args of the class'''
        self.keyword = keyword
        if keyword in ('input', 'copyright', 'credits', 'help', 'print', 'display', 'license', 'dir'):
            # We don't wanna try to evaulate these, cause they cause anoying outputs
            # When we use filter on a map of CallableKeywords and list(keywords) if the keyword is in the above 
            # tuple the mapitem will then be a bool : False
            # filter filters away false
            return False
        try:
            self.function = eval(keyword) # wrapp the function
            self.result = self.function(*self.args) # call the function with *args
            return self # return instance of self
        except Exception as e: # except any errors
            self.error = e # set the error message to self.error attribute
            self._parseError() # TODO: implement an error parser
            self.accept = not self.accept # A bool to know if the function accepts the arguments
            # lets keep this bool value, in case we need it
            return self # return instance of self
        
    def getKeyword(self):
        return self.keyword
        
    def error(self):
        __doc__ = '''If an error is raise when calling the class object as a function, 
        it can be retrieved by this method'''
        return self.error
    
    def _parseError(self):
        # TODO: build an error parser
        __doc__ = '''_parseError will try to figure out how to call the keyword function, with correct parameters'''
        pass
    
    def result(self):
        __doc__ = ''' return the result from the function call '''
        return self.result
    
    def __str__(self):
        __doc__ = ''' return the class name and input arguments '''
        return f'{self.name}{self.args}({self.keyword})'
    
    def __repr__(self):
        __doc__ = ''' return the class name and input arguments '''
        return f'{self.name}{self.args}({self.keyword})'
    
    def __dict__(self):
        return {
                'args'    : self.args,
                'accept'  : self.accept,
                'keyword' : self.keyword,
                'function': self.function,
                'error'   : self.error,
                'result'  : self.result,
                'name'    : self.name
               }
    
    def __hash__(self):
        return hash(id(self))
    
    
    def close(self):
        self.keyword = 'Not initialized'
        self.function = 'Not initialized'
        self.error = 'Not raised'
        self.result = None
        self.name = self.init_name()
        del self.args
```

```python
def searchEngine1(word, module):
    try:
        mod = sys.modules[module]
    except:
        return f'No module named: {module}'
    results = {module: []}
    for attr in dir(mod):
        #print(attr)
        if word in attr:
            results[module].append(attr)
    return results
```

```python
def searchEngine3(word):
    results = {}
    for modulename, module in sys.modules.items():
        results[modulename] = []
        for attr in dir(module):
            if word.lower() in attr.lower():
                results[modulename].append(attr)
        if not results[modulename]:
            results.pop(modulename)
    return results
```

```python
def searchEngine4(word, accuracy=0.80):
    results = {}
    for modulename, module in sys.modules.items():
        results[modulename] = []
        for attr in dir(module):
            attr_length = len(attr)
            word_length = len(word)
            num_shifts = attr_length - word_length # we want to match every shift example:
            # attr = "ThisIsAModule"
            # word = "IsAModule"
            # I wil refer SequenceMatcher as match and without the first argument
            # we will match liek this match("ThisIsAModule"[0:len(IsAModule)], "IsAModule")
            # we will match liek this match("ThisIsAModule"[1:len(IsAModule)], "IsAModule")
            # we will match liek this match("ThisIsAModule"[len("ThisIsAModule") - len("IsAModule") : len("IsAModule")], "IsAModule")
            start = 0
            all_ratio = []
            
            while num_shifts > 0:
                match_attr = attr[start:][:word_length] # or [start:word_length + start] #with strings you can use as many __getitem__ as you want ex 'a'[0][0][0] = 'a'
                all_ratio.append(SequenceMatcher(None, word, match_attr).quick_ratio()) # lets use the quick_ratio, as real_quick_ratio looked a little off
                start += 1
                num_shifts -= 1
            else:
                all_ratio.append(SequenceMatcher(None, word, attr).quick_ratio())
                
            best_match = max(all_ratio)
            
            if best_match >= accuracy:
                #print(best_match)
                results[modulename].append(attr)
                
        if not results[modulename]:
            results.pop(modulename)
    return results
```

```python
def search_site_packages(word):
    results = []
    paths = [pathlib.Path(path) for path in filter(lambda x:x.endswith('site-packages'), sys.path)]
    for path in paths:
        for module in path.glob('**/*.py'):
            name = str(module)
            with module.open() as f: 
                try:
                    line = f.readline()
                    while line:
                        if word in line:
                            results.append(name)
                            break
                        line = f.readline()
                except:
                    pass
    return results
        
```
